## RFM Feature Engineering

Recency, Frequency, and Monetary (RFM) metrics are calculated using leakage-free transaction data to capture customer purchasing behavior.

```python
snapshot_date = churn["snapshot_date"].max()

snapshot_date = pd.to_datetime(snapshot_date)

print("Snapshot Date:", snapshot_date)
```

### Recency Definition

Recency is calculated as the number of days since a customer's last purchase prior to the snapshot date.

Although website visit activity is available through web_events, purchase recency was selected because it is directly tied to revenue generation and follows the standard RFM framework used in customer analytics.

Website engagement behaviour is captured separately through sessions_30d and campaign engagement metrics.

```python
last_order = (
    orders_filtered
    .groupby("customer_id")["order_date"]
    .max()
    .reset_index()
)

last_order["recency"] = (
    snapshot_date -
    last_order["order_date"]
).dt.days

last_order.head()
```

```python
print("Customers:", len(last_order))

print(
    "Minimum Recency:",
    last_order["recency"].min()
)

print(
    "Maximum Recency:",
    last_order["recency"].max()
)
```

### Frequency
Total number of purchases made before the snapshot date.

```python
frequency = (
    orders_filtered
    .groupby("customer_id")
    .size()
    .reset_index(name="frequency")
)

frequency.head()
```

```python
frequency["frequency"].describe()
```

### Monetary
Total revenue generated by each customer.

```python
monetary = (
    orders_filtered
    .groupby("customer_id")["gross_amount"]
    .sum()
    .reset_index()
)

monetary.columns = [
    "customer_id",
    "monetary"
]

monetary.head()
```

```python
monetary["monetary"].describe()
```

### Monetary Value Definition

Monetary value is calculated using gross transaction amounts from valid pre-snapshot orders.

Return behaviour is analysed separately through return_rate and therefore is not deducted from monetary value during RFM calculation.

### RFM Table
Combine Recency, Frequency, and Monetary metrics into a single customer-level dataset.

```python
rfm = (
    last_order[
        ["customer_id", "recency"]
    ]
    .merge(
        frequency,
        on="customer_id"
    )
    .merge(
        monetary,
        on="customer_id"
    )
)

rfm.head()
```

```python
print("Rows:", len(rfm))

print()

rfm.describe()
```

```python
rfm.to_csv(
    "../outputs/tables/rfm_features.csv",
    index=False
)
```

### RFM Feature Design Decisions

Recency is calculated as the number of days since a customer's most recent purchase before the snapshot date.

Although website visit activity is available through web_events, purchase recency was selected because it is directly tied to revenue generation and follows the standard RFM framework.

Website engagement behaviour is captured separately through sessions_30d, email engagement, and campaign interaction features.

```python
print(rfm.columns.tolist())
```

```python
# RFM Score Boundaries

r_boundaries = rfm["recency"].quantile([0.2, 0.4, 0.6, 0.8])
f_boundaries = rfm["frequency"].quantile([0.2, 0.4, 0.6, 0.8])
m_boundaries = rfm["monetary"].quantile([0.2, 0.4, 0.6, 0.8])

boundary_table = pd.DataFrame({
    "Recency": r_boundaries,
    "Frequency": f_boundaries,
    "Monetary": m_boundaries
})

boundary_table
```

### RFM Score Interpretation

The quintile boundaries above determine score assignment.

- R=5 represents the most recent customers.
- F=5 represents the most frequent purchasers.
- M=5 represents the highest-spending customers.

Providing raw-value thresholds improves transparency and allows business stakeholders to audit score assignments.

## RFM Scoring

Customers are scored from 1–5 for Recency, Frequency, and Monetary value using quintile-based ranking.

```python
rfm["R_score"] = pd.qcut(
    rfm["recency"],
    5,
    labels=[5,4,3,2,1]
)
```

```python
rfm["F_score"] = pd.qcut(
    rfm["frequency"].rank(
        method="first"
    ),
    5,
    labels=[1,2,3,4,5]
)
```

```python
rfm["M_score"] = pd.qcut(
    rfm["monetary"],
    5,
    labels=[1,2,3,4,5]
)
```

```python
rfm["R_score"] = rfm["R_score"].astype(int)

rfm["F_score"] = rfm["F_score"].astype(int)

rfm["M_score"] = rfm["M_score"].astype(int)
```

```python
rfm["RFM_SCORE"] = (
    rfm["R_score"].astype(str)
    +
    rfm["F_score"].astype(str)
    +
    rfm["M_score"].astype(str)
)
```

## Customer Segmentation

Customers are segmented using RFM scores and additional behavioral indicators such as support interactions, discount dependency, and customer sentiment.

The segmentation logic is designed to identify customers with different retention needs and business value profiles.

```python
master = master.merge(
    rfm[
        [
            "customer_id",
            "R_score",
            "F_score",
            "M_score",
            "RFM_SCORE"
        ]
    ],
    on="customer_id",
    how="left"
)

master.shape
```

### Segment Rule Priority

Customers may satisfy multiple segment conditions simultaneously. To avoid ambiguity, segment rules are evaluated in business-priority order, with customer value and dissatisfaction indicators taking precedence over general behavioural classifications.

```python
# Campaign Engagement Feature

campaign_features = campaigns[
    [
        "customer_id",
        "last_campaign_received",
        "manual_priority_bucket"
    ]
].copy()

campaign_features["campaign_engaged"] = (
    campaign_features["last_campaign_received"]
    .ne("none")
    .astype(int)
)

master = master.merge(
    campaign_features[
        ["customer_id", "campaign_engaged"]
    ],
    on="customer_id",
    how="left"
)

master["campaign_engaged"] = (
    master["campaign_engaged"]
    .fillna(0)
    .astype(int)
)

print(master[["customer_id", "campaign_engaged"]].head())
```

```python
def assign_segment(row):

    # Champions
    if (
        row["R_score"] >= 4
        and row["F_score"] >= 4
        and row["M_score"] >= 4
    ):
        return "Champions"

    # Loyal Customers
    elif (
        row["R_score"] >= 3
        and row["F_score"] >= 4
    ):
        return "Loyal Customers"

    # Potential Loyalists
    elif (
        row["R_score"] >= 4
        and row["F_score"] >= 2
    ):
        return "Potential Loyalists"

    # New Customers
    elif (
        row["R_score"] >= 4
        and row["F_score"] == 1
    ):
        return "New Customers"

    # High Value Unhappy
    elif (
        row["M_score"] >= 4
        and row["ticket_count"] >= 2
        and row["avg_sentiment"] < 0
    ):
        return "High Value Unhappy"

    # Discount Sensitive
    elif (
        row["avg_discount"] >= 0.35
    ):
        return "Discount Sensitive"

    # At Risk
    elif (
        row["R_score"] <= 2
        and row["M_score"] >= 3
    ):
        return "At Risk"

    # Dormant
    elif (
        row["R_score"] == 1
        and row["F_score"] <= 2
        and row["sessions_30d"] == 0
        and row["campaign_engaged"] == 0
    ):
        return "Dormant"

    # Promising
    elif (
        row["R_score"] == 3
        and row["F_score"] >= 2
    ):
        return "Promising"

    # Regular Customers
    elif (
        row["R_score"] >= 2
        and row["R_score"] <= 4
        and row["F_score"] >= 2
    ):
        return "Regular Customers"

    # Fallback
    else:
        return "Regular Customers"
```

```python
# Sensitivity Analysis: At-Risk Threshold

print("At-Risk Segment Sensitivity Analysis")

for r_thresh in [1, 2, 3]:
    
    count = master[
        (master["R_score"] <= r_thresh)
        & (master["M_score"] >= 3)
    ].shape[0]
    
    print(
        f"R_score <= {r_thresh}: {count} customers"
    )
```

### At-Risk Threshold Sensitivity

The At-Risk segment was tested using multiple Recency score thresholds while keeping the Monetary score requirement fixed (`M_score >= 3`).

| Rule | Customers |
|------|----------:|
| R_score <= 1 | 278 |
| R_score <= 2 | 570 |
| R_score <= 3 | 882 |

The selected threshold (`R_score <= 2`) balances customer coverage and segment precision. A stricter threshold would miss potentially churning customers, while a broader threshold would substantially increase segment size and overlap with more engaged customer groups.

## Segmentation Logic

Customer segments were created using RFM scores and additional behavioral signals such as support tickets, sentiment, discount usage, and customer activity.

| Segment | Key Criteria |
|----------|-------------|
| Champions | High R, F, and M scores |
| Loyal Customers | High frequency and recent purchases |
| Potential Loyalists | Recent customers with growing engagement |
| New Customers | Recent first-time buyers |
| Promising | Moderate activity with growth potential |
| High Value Unhappy | High spenders with negative support experience |
| Discount Sensitive | High discount dependency |
| At Risk | Valuable customers with declining activity |
| Dormant | Long inactive customers |
| Regular Customers | Customers not matching other segments |

```python
master["segment_name"] = (
    master.apply(
        assign_segment,
        axis=1
    )
)

master[
    [
        "customer_id",
        "segment_name"
    ]
].head()
```

```python
segment_summary = (
    master["segment_name"]
    .value_counts()
    .reset_index()
)

segment_summary.columns = [
    "segment_name",
    "customer_count"
]

segment_summary["percentage"] = (
    segment_summary["customer_count"]
    / segment_summary["customer_count"].sum()
    * 100
).round(2)

segment_summary
```

## Budget Prioritization

To prioritize retention spending, a simple business-driven priority score was calculated using:

Priority Score = Customer Count × Average Revenue × Churn Rate

This approach favors segments that:

- contain many customers,
- generate high revenue,
- exhibit elevated churn risk.

Segments with the highest scores should receive retention investment first because they offer the greatest potential business impact.
### Budget Assumption

For prioritization purposes, a hypothetical retention budget of ₹500,000 is assumed.

Priority scores are calculated using customer count, average revenue, and observed churn rate. Segments with higher scores are expected to deliver greater revenue protection if retention efforts are successful.

```python
budget_priority = (
    master.groupby("segment_name")
    .agg(
        customer_count=("customer_id", "count"),
        avg_revenue=("monetary", "mean"),
        churn_rate=("churn_next_60d", "mean")
    )
    .reset_index()
)

budget_priority["priority_score"] = (
    budget_priority["customer_count"]
    * budget_priority["avg_revenue"]
    * budget_priority["churn_rate"]
)

budget_priority = budget_priority.sort_values(
    by="priority_score",
    ascending=False
)

budget_priority
```

### Budget Allocation Guidance

Retention investment should be prioritised using the segment priority score, which combines customer volume, average revenue, and observed churn risk.

Highest-priority segments:

1. At Risk
2. High Value Unhappy
3. Discount Sensitive
4. Loyal Customers

These segments represent the largest revenue recovery opportunities and are expected to deliver the greatest retention impact per campaign cycle.

### Intervention Measurement Framework

To evaluate retention campaign effectiveness, 10–15% of customers in each segment should be held out as a control group.

Comparing retention outcomes between treatment and control groups enables measurement of incremental retention lift and helps determine whether campaign interventions generate meaningful business impact.

```python
budget_priority.to_csv(
    "../outputs/tables/budget_priority.csv",
    index=False
)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=budget_priority,
    x="priority_score",
    y="segment_name"
)

plt.title("Retention Budget Priority by Segment")

plt.tight_layout()

plt.savefig(
    "../outputs/charts/budget_priority.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
